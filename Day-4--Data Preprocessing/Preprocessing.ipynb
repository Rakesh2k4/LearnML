```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data.csv')
df
```

```python
# Finding number of null values
df.isnull().sum()
```

```python
# Handling numerical columns by filling them with mean
df['Age'].fillna(df['Age'].sum(),inplace=True)
df['Salary'].fillna(df['Salary'].mean(),inplace=True)
```

```python
# Checking null values
df.isnull().sum()
```

```python
# We need to convert categorical data to numerical 
# We need to convert Purchased column by Label encoding
df['Purchased'] = df['Purchased'].map({'Yes':1,'No':0})
```

```python
# We need to convert Country column by One-Hot encoding
df = pd.get_dummies(df,columns=['Country'],dtype=int)
```
