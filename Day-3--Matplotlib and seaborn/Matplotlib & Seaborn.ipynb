```python
# importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing the dataset
df = pd.read_csv('HistoricalData_1755100883008.csv')
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
# Histogram
plt.hist(df['Volume'],bins=5,color='blue',alpha=0.5)
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.title('Volume v/s Frequency')
plt.show()
```

```python
#barplot
plt.bar(df['Date'],df['Close/Last'],color='green')
plt.show()
```

```python
# line plot
plt.plot(df['Date'],df['Close/Last'],label='AAPL')
plt.legend()
plt.gcf().autofmt_xdate()
plt.show()
```

```python
# histogram
sns.histplot(data = df['Volume'],bins=10,color='blue')
plt.show()
```

```python
# barplot
sns.barplot(x='Date',y='Close/Last',data = df,color='green')
plt.show()
```

```python
sns.lineplot(x='Date',y='Close/Last',data = df)
plt.show()
```

```python
sns.scatterplot(x='Date',y='Close/Last',data = df)
plt.show()
```

```python
sns.boxplot(x='Volume',data=df)
```

```python
sns.violinplot(y="Volume",data=df)
```
